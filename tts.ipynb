```python
import pyttsx3
import pandas as pd
import os

# Load your dataset
df = pd.read_csv("pidgin_tts_dataset.csv")

# Create folder for audio
os.makedirs("audio_dataset", exist_ok=True)

# Initialize engine
engine = pyttsx3.init()

# Optional: Adjust voice settings
engine.setProperty("rate", 150)     # Speed (words per minute)
engine.setProperty("volume", 1.0)   # Volume (0.0 to 1.0)

# Select a voice (depends on your system)
voices = engine.getProperty("voices")
for i, voice in enumerate(voices):
    print(f"Voice {i}: {voice.name} - {voice.id}")

# Example: pick a different voice if available
engine.setProperty("voice", voices[0].id)  # Change index if you want another voice

# Generate audio files
for i, row in df.iterrows():
    text = str(row["Sentence"])  # Make sure column matches
    filename = f"audio_dataset/{i}.mp3"
    engine.save_to_file(text, filename)
    print(f"Saved {filename}")

engine.runAndWait()  # Process all queued audio at once

print("✅ Audio dataset created!")
```
