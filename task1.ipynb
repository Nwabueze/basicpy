```python
def count(lst):
    even = 0
    odd = 0

    for i in lst:
        if i % 2 == 0:
            even += 1
        else:
            odd += 1

    return even, odd

lst = [20, 25, 14, 19, 16, 24, 28, 47, 26]
even, odd = count(lst)
print("Even : {} and Odd : {}".format(even, odd))
```

```python
# Working with Python lists

# Initialize a list of numbers
nums = [25, 12, 36, 95, 14]

# Accessing elements
print(nums[4])      # prints the element at index 4 → 14
print(nums[2:])     # prints elements from index 2 to the end → [36, 95, 14]
print(nums[-1])     # prints last element → 14
print(nums[-5])     # prints first element (since -5 is same as index 0) → 25

# List of names
names = ["ann", "kylla", "sammy"]
print(names)        # prints the names list

# Mixed list with string, float, integer
values = ["name", 9.8, 42]

# Nested list (list inside another list)
mil = [nums, names]
print(mil)          # prints nested list containing nums and names

# Add a new element to the end of nums
nums.append(45)
print(nums)         # nums now has 45 added at the end

# Insert element at a specific position
nums.insert(2, 77)  
print(nums)         # 77 is inserted at index 2

# Remove an element by value
nums.remove(25)     
print(nums)

# Delete elements by index range
del nums[2:]
print(nums)         # keeps only first 2 elements

# Extend the list by adding multiple elements
nums.extend([100, 200, 300])
print(nums)         # nums now has extra 3 elements at the end

# Built-in functions on lists
print(min(nums))    # prints minimum value in nums
print(max(nums))    # prints maximum value in nums
print(sum(nums))    # prints sum of all values in nums

# Sorting the list in ascending order
nums.sort()
print(nums)         # prints sorted list
```

```python
# Demonstrating Python variables, memory IDs, and immutability of objects

# Assign integer value 5 to variable 'num'
num = 5
# Get the unique memory address (ID) of 'num'
id(num)

# Assign string value "John" to variable 'name'
name = "John"
# Print the memory address (ID) of 'name'
print(id(name))

# Assign integer 10 to variable 'a'
a = 10
# Assign the value of 'a' to 'b' (both now reference the same object 10)
b = a
print(a)   # Output: 10
print(b)   # Output: 10

# Get memory address (ID) of 'a' and 'b'
id(a)
id(b)

# Get memory address (ID) of integer literal 10
id(10)

# Assign 10 to 'k' (it will point to the same object 10 as 'a' and 'b' initially)
k = 10
id(k)

# Reassign 'a' to a new integer 9 (immutability: new object is created, new ID)
a = 9
id(a)

# 'b' is still pointing to 10
id(b)

# Now reassign 'k' to the value of 'a' (so k now points to 9)
k = a
id(k)

# Reassign 'b' to 8 (b now points to a new integer object)
b = 8

# By convention, constants in Python are written in ALL CAPS (but still mutable)
PI = 3.14   # Assign float 3.14 to PI
PI          # Evaluates to 3.14

# Reassign PI to 3.15 (allowed, since Python does not enforce constants)
PI = 3.15

# Check the type of PI (will output <class 'float'>)
type(PI)
```

```python
# Python dictionary

data = {1: "John", 2: "Doe", 3: "Anna"}

print(data)

print(data[1])

# Return the value for key if key is in the dictionary, else default
print(data.get(2))
print(data.get(5, "Not Found"))

# merging two array as dictionary
keys = ["key1", "key2", "key3"]
vals = ["value1", "value2", "value3"]

merged_dict = dict(zip(keys, vals))

print(merged_dict)

# Assign a new key value
merged_dict["key4"] = "value4"
merged_dict["key5"] = "value5"

print(merged_dict)

# delete a key
del merged_dict["key1"]

print(merged_dict)

# nested dictionary of mixed single values, list, and other dictionaries
nested_dict = {
    "key1": "value1",
    "key2": ["value2a", "value2b"],
    "key3": {"subkey1": "subvalue1", "subkey2": "subvalue2"}
}

print(nested_dict)
```

```python
# dict_demo.py
# A complete demo of Python dictionaries and their operations

def main():
    # 1. Creating dictionaries
    student = {
        "name": "Alice",
        "age": 22,
        "major": "Computer Science"
    }
    print("Original dictionary:", student)

    # 2. Accessing values
    print("Name:", student["name"])
    print("Age:", student.get("age"))   # safer way

    # 3. Adding new key-value pairs
    student["GPA"] = 3.8
    print("\nAfter adding GPA:", student)

    # 4. Updating values
    student["age"] = 23
    print("After updating age:", student)

    # 5. Removing keys
    gpa = student.pop("GPA")  # removes and returns value
    print(f"Removed GPA ({gpa}):", student)

    # 6. Checking if a key exists
    if "major" in student:
        print("Key 'major' exists with value:", student["major"])

    # 7. Iterating over dictionary
    print("\nIterating keys:")
    for key in student:
        print(key, "->", student[key])

    print("\nIterating items:")
    for key, value in student.items():
        print(key, ":", value)

    # 8. Dictionary methods
    print("\nKeys:", list(student.keys()))
    print("Values:", list(student.values()))
    print("Items:", list(student.items()))

    # 9. Nested dictionaries
    classroom = {
        "student1": {"name": "Bob", "age": 20},
        "student2": {"name": "Carol", "age": 21}
    }
    print("\nNested dictionary example:", classroom)
    print("Student1's name:", classroom["student1"]["name"])

    # 10. Dictionary comprehension
    squares = {x: x**2 for x in range(5)}
    print("\nDictionary comprehension (squares):", squares)

    # convert list to dict
    names = ["Alice", "Bob", "Carol"]
    lengths = {name: len(name) for name in names}
    print(lengths)

    # Filtering with condition
    even_squares = {x: x**2 for x in range(10) if x % 2 == 0}
    print(even_squares)

if __name__ == "__main__":
    main()
```

```python
# For loop


# 1. Simple for loop with a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

# 2. For loop with range
for i in range(5):  # 0 to 4
    print("Number:", i)

# 3. Loop through a string
for ch in "hello":
    print(ch)

# 4. Using break
for num in range(10):
    if num == 5:
        break
    print("Breaking at:", num)

# 5. Using continue
for num in range(5):
    if num == 2:
        continue
    print("Skipping 2:", num)

# 6. Nested loops
colors = ["red", "blue"]
shapes = ["circle", "square"]
for color in colors:
    for shape in shapes:
        print(color, shape)


# 7. Loop with list comprehension (basic)
squares = [x**2 for x in range(5)]
print("Squares:", squares)



```

```python
# -----------------------------------
# Advanced Python For Loop Examples
# -----------------------------------

# 1. Looping with enumerate
fruits = ["apple", "banana", "cherry"]
for index, fruit in enumerate(fruits):
    print(index, fruit)

# 2. Looping with zip (parallel iteration)
names = ["Alice", "Bob", "Carol"]
scores = [85, 92, 78]
for name, score in zip(names, scores):
    print(f"{name} scored {score}")

# 3. Dictionary iteration
person = {"name": "John", "age": 25, "city": "New York"}
for key, value in person.items():
    print(key, ":", value)

# 4. Using range with step
for i in range(0, 20, 5):
    print("Step of 5:", i)

# 5. Looping backwards
for i in reversed(range(5)):
    print("Reversed:", i)

# 6. Loop with any/all
nums = [2, 4, 6, 8]
if all(n % 2 == 0 for n in nums):
    print("All numbers are even")

# 7. Generator expression inside loop
gen = (x**2 for x in range(5))
for val in gen:
    print("Generated square:", val)

# 8. Loop with try/except inside
for val in ["10", "20", "x", "40"]:
    try:
        num = int(val)
        print("Converted:", num)
    except ValueError:
        print("Invalid number:", val)

# 9. Complex comprehension with condition
even_squares = {x: x**2 for x in range(10) if x % 2 == 0}
print("Even Squares Dict:", even_squares)

# 10. For + else with break condition
for i in range(10):
    if i == 7:
        print("Found 7, breaking loop")
        break
else:
    print("Did not break the loop")
```

```python
# Basic while loop operations in Python

# 1. Simple counter
i = 1
while i <= 5:
    print("Count:", i)
    i += 1

# 2. While loop with break
i = 1
while i <= 10:
    if i == 5:
        print("Breaking at i =", i)
        break
    print(i)
    i += 1

# 3. While loop with continue
i = 0
while i < 5:
    i += 1
    if i == 3:
        print("Skipping 3")
        continue
    print(i)

# 4. While loop with else
i = 1
while i <= 3:
    print("i =", i)
    i += 1
else:
    print("Loop finished without break")

# 5. Infinite loop example (commented out for safety)
# while True:
#     print("This will run forever!")
#     break   # normally you need a break to stop it
```

```python
# Advanced while loop operations in Python

# 1. Generating a multiplication table using while loop
num = 7
i = 1
while i <= 10:
    print(f"{num} x {i} = {num * i}")
    i += 1

# 2. Summing numbers until user enters 0
total = 0
num = int(input("Enter a number (0 to stop): "))
while num != 0:
    total += num
    num = int(input("Enter another number (0 to stop): "))
print("Total sum =", total)

# 3. Reverse a number using while loop
n = 12345
rev = 0
while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n //= 10
print("Reversed number:", rev)

# 4. Factorial using while loop
n = 5
fact = 1
while n > 0:
    fact *= n
    n -= 1
print("Factorial:", fact)

# 5. Fibonacci sequence with while loop
a, b = 0, 1
count = 0
while count < 10:
    print(a, end=" ")
    a, b = b, a + b
    count += 1
print()

# 6. Nested while loops (printing a pattern)
rows = 5
i = 1
while i <= rows:
    j = 1
    while j <= i:
        print("*", end=" ")
        j += 1
    print()
    i += 1
```
